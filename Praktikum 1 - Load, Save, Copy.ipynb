### Import Dependency

```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
```

### Load Image

```python
img = imageio.imread("woman.jpg")
```

### Get Image Properties

```python
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
```

### Show Image

```python
plt.imshow(img)
plt.title("Load Image")
plt.show()
```

### Copy Image

#### Create image variabel

```python
img_copy = np.zeros(img.shape, img.dtype)
```

#### Copy image algorithm

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_copy[y][x][c] = img[y][x][c]
```

#### Show image

```python
plt.imshow(img_copy)
plt.title("Copy")
plt.show()
```

### Save Image

```python
plt.imsave("saved-copy.jpg", img_copy)
```
