### Import dependency

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

### Membuat Gambar Berwarna Merah

#### Membuat variabel untuk gambar berukuran 256 pixel x 256 pixel

```python
img_red = np.zeros((256, 256, 3), dtype=np.uint8)
```

#### Set pixel with [255, 0, 0] or red

```python
for y in range(0, 256):
    for x in range(0, 256):
        for c in range(0, 3):
            if c == 0:
                img_red[y][x][c] = 255
            else:
                img_red[y][x][c] = 0
```

#### Test image

```python
plt.imshow(img_red)
plt.title("Red")
plt.show()
```

### Create Indonesian's Flag

#### Create variabel

```python
img_IND_flag = np.zeros((640, 1280, 3), dtype=np.uint8)
```

#### Set the pixel

```python
for y in range(0, 640):
    for x in range(0, 1280):
        for c in range(0, 3):
            if y >= 320:
                img_IND_flag[y][x][c] = 255 # set semua channel menjadi 255 atau putih [255,255,255]
            else:
                if c == 0:
                    img_IND_flag[y][x][c] = 255
                else:
                    img_IND_flag[y][x][c] = 0
```

#### Test image

```python
plt.imshow(img_IND_flag)
plt.title("Bendera Indonesia")
plt.show()
```

### Flip Image

#### Membaca gambar

```python
img = imageio.imread("woman.jpg")
```

#### Mendapatkan resolusi dan type dari gambar

```python
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype
```

#### Membuat variabel dengan resolusi dan tipe yang sama seperti gambar

```python
img_flip_horizontal = np.zeros(img.shape, img_type)
img_flip_vertical = np.zeros(img.shape, img_type)
```

#### Membalik gambar secara horizontal

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c]
```

#### Membalik gambar secara vertical

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c]
```

#### Menampilkan hasil balik gambar

```python
plt.imshow(img_flip_horizontal)
plt.title("Flip Horizontal")
plt.show()
plt.imshow(img_flip_vertical)
plt.title("Flip Vertical")
plt.show()
```
