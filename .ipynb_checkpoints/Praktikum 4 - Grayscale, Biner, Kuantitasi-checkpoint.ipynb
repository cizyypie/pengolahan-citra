## Import dependency

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

## Membaca gambar

```python
img = imageio.imread("woman.jpg")
```

## Mendapatkan resolusi dan type dari gambar

```python
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype
```

------------------------------------------------

## Grayscale

### Membuat variabel img_gray untuk menampung hasil

```python
img_gray = np.zeros(img.shape, dtype=np.uint8)
```

### Menghitung rata2 RGB, lalu memplot hasilnya pada setiap nilai R, G, dan B dalam pixel

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        gray = (int(red) + int(green) + int(blue)) / 3
        img_gray[y][x] = (gray, gray, gray)
```

### Menampilkan Hasil Grayscale

```python
plt.imshow(img_gray)
plt.title("Grayscale")
plt.show()
```

-------------------------------------

## Biner

### Membuat variabel img_biner untuk menampung hasil

```python
img_biner = np.zeros(img.shape, dtype=np.uint8)
```

### Menghitung rata2 RGB, dan kemudian membagi dua bagian dalam nilai rata2, jika kurang dari setengah (128) maka menjadi 0 jika lebih maka menjadi penuh (255), lalu memplot hasilnya pada setiap nilai R, G, dan B dalam setiap pixel

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        gray = (int(red) + int(green) + int(blue))/3
        if gray < 128:
            gray = 0
        else:
            gray = 255
        img_biner[y][x] = (gray, gray, gray)
```

### Menampilkan Hasil Biner

```python
plt.imshow(img_biner)
plt.title("Biner")
plt.show()
```

-------------------------

## Kuantitasi

### Normal

```python
plt.imshow(img_gray)
plt.title("Normal Gray")
plt.show()
```

### With Thresholding 16

```python
th = int(256/16)
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        wgray = (int(red) + int(green) + int(blue)) / 3
        xgray = th * int(wgray/th)
        img_gray[y][x] = (xgray, xgray, xgray)
```

```python
plt.imshow(img_gray)
plt.title("Gray TH 16")
plt.show()
```

### With Thresholding 4

```python
th = int(256/4)
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        wgray = (int(red) + int(green) + int(blue)) / 3
        xgray = th * int(wgray/th)
        img_gray[y][x] = (xgray, xgray, xgray)
```

```python
plt.imshow(img_gray)
plt.title("Gray TH 4")
plt.show()
```

### With Thresholding 2

```python
th = int(256/2)
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        wgray = (int(red) + int(green) + int(blue)) / 3
        xgray = th * int(wgray/th)
        img_gray[y][x] = (xgray, xgray, xgray)
```

```python
plt.imshow(img_gray)
plt.title("Gray TH 2")
plt.show()
```
